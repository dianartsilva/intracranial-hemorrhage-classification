The following script was used to build a CSV file from the original file and more suitable for our task. 

The new CSV file is going to have 3 columns:
*   "ID": with the name of the image
*   "presence": with the labels "hemorrhage" and "no hemorrhage"
*   "type": with the labels "epidural", "intraventricular", "intraparenchymal", "subarachnoide", "subdural" and "no hemorrhage" (this label only exists for the images which presence is "no hemorrhage" too)

```python
import os
import shutil
import pandas as pd

# Folder paths
origin_original = '/content/drive/Shareddrives/Projeto DACO/small dataset/original/'
target_original = '/content/drive/Shareddrives/Projeto DACO/single label dataset/original/'

origin_window = '/content/drive/Shareddrives/Projeto DACO/small dataset/windowing/'
target_window = '/content/drive/Shareddrives/Projeto DACO/single label dataset/windowing/'

data = pd.read_csv('/content/drive/Shareddrives/Projeto DACO/stats - png small dataset/small_dataset_nomulti.csv')

files = set(data['ID'])
# Fetching all the files to directory
for file_name in files:
  print(file_name)
  shutil.copy(origin_original+file_name+".png", target_original+file_name+".png")
  shutil.copy(origin_window+file_name+".png", target_window+file_name+".png")
```

```python
import numpy as np
import pandas as pd

csv = pd.read_csv('/content/drive/Shareddrives/Projeto DACO/single label dataset/small_dataset_label.csv')
ids = np.unique(csv['ID'])

labels = []

for i in ids:
  label = [i for i in csv[csv['ID']==i]['label']]
  if label[-1]==' none':
          labels.append('no hemorrhage')
  else:
    label =  [int(i) for i in label[:-2]]
    hem =[i for i in csv[csv['ID']==i]['type']][:-2]
    label = hem[label.index(1)]

    labels.append(label)

info= pd.DataFrame({"ID": ids, "label": labels})
info.to_csv('/content/drive/Shareddrives/Projeto DACO/single label dataset/single_label.csv', index=False, header=True)
```

```python
import numpy as np
import pandas as pd

csv = pd.read_csv('/content/drive/Shareddrives/Projeto DACO/single label dataset/single_label.csv')
ids = np.unique(csv['ID'])

labels = []

for i in ids:
  label = [i for i in csv[csv['ID']==i]['label']]
  if (label[-1]=='no hemorrhage'):
          labels.append('no hemorrhage')
  else:
    labels.append('hemorrhage')

info= pd.DataFrame({"ID": ids, "label": labels})
info.to_csv('/content/drive/Shareddrives/Projeto DACO/single label dataset/hemmorhage_presence.csv', index=False, header=True)
```
